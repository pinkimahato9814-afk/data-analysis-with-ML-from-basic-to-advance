# Fill Null Records

```python
# fillna()
# permanent change -> inplace=True
# mean, median, mode, custom record
```

```python
mean_commission = hr['COMMISSION_PCT'].mean()
mean_commission
```

```python
# NumPy -> Numerical Python
# Pandas
```

```python
# hr['COMMISSION_PCT'].median()
```

```python
# round -> int / float
round_mean_com = round(mean_commission, 2)
round_mean_com
```

```python
# hr['COMMISSION_PCT'].fillna(round_mean_com, inplace=True)
hr['COMMISSION_PCT'] = hr['COMMISSION_PCT'].fillna(round_mean_com)
```

```python
# format -> string
# format_com = f'${mean_commission:,.2f}'
```

```python
mgr_mode = hr['MANAGER_ID'].mode()[0]
mgr_mode
```

```python
dep_mode = hr['DEPARTMENT_ID'].mode()[0]
dep_mode
```

```python
hr['MANAGER_ID'] = hr['MANAGER_ID'].fillna(mgr_mode)
hr['DEPARTMENT_ID'] = hr['DEPARTMENT_ID'].fillna(dep_mode)
```

```python
# Replace() -> String Function
# replace(find, replace_value, regex=False)
hr['PHONE_NUMBER'] = hr['PHONE_NUMBER'].str.replace('.','-', regex=False)
```

# Strip()

```python
# Use of strip is removing white space.
# Strip() -> remove white space in both side.
# lstrip() -> remove white space in left side only.
# rstrip() -> remove white space in right side only.
```

```python
for strip_col in columns:
    df[strip_col] = df[strip_col].str.strip()
```

# Label Encoding

```python
# map() -> dictionary
hl_dict = {
    'No': 'No',
    'Yes': 'Yes',
    'no': 'No',
    'yes': 'Yes'
}
df['HispanicLatino'] = df['HispanicLatino'].map(hl_dict)
```
