# meshgrid and linespace function understand by using graph(this is  for extra knowledge )

```python



#linspace → “give me points along a line” function->linespace(start,stop,how many number you want to generate)
#meshgrid → “turn those lines into a full grid”-> A,B = np.meshgrid(a,b)
# A grid is a structured set of points arranged in rows and columns.

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contourf(X, Y, Z)
plt.colorbar()
plt.show()
```

```python
# loc and iloc in pandas,specially for slicing

import pandas as pd

data = {
    'name': ['A', 'B', 'C', 'D'],
    'age': [20, 25, 30, 35],
    'salary': [1000, 2000, 3000, 4000],
    'address':["k","p","r","t"],
    "qualification": ['master','phd','bachelor','nursing']
}

df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'])
df

```

1. loc (Label-based slicing)
df.loc[row_label, column_label]


```python
# df.loc[row_label, column_label] --> row and column level slicing
# key rules for loc - Uses labels (names) and includes the last value
# loc is mainly used for filtering data with conditions
df
#df.loc['d':'address']# why this is not work
df.loc['a':'address'] # only this is work
```

```python
#df.loc['a':'c',['name' ,'age']] and 
df.loc['a':'c',['name' ,'age','address'] ]
#df.iloc[[0, 2]]
# df.iloc['a','age']
```

```python
df.loc['a':'c', ['name', 'salary','address']]
```

2. iloc (Position-based slicing)
df.iloc[row_index, column_index]

```python
#Example 1: Single row
df.iloc[3]
```

```python
# Multiple rows
df.iloc[[0, 1]]
```

```python
# Example 3: Slicing rows
df.iloc[0:3]
```

```python
# Example 4: Rows + columns
df.iloc[0:3, 0:2]
```
