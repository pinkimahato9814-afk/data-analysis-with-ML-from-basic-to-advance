## if we not required all details between range , IF you wnat specific column then we use loc function
# df.loc[df['filter_column'].between(range),[number of column which we want to extract]       ]

```python
#Q-> find out first_name ,last_name and salary of employee whose salary range between 5000 to 10000
df.loc[df['SALARY'].between(5000,10000),['FIRST_NAME','LAST_NAME','SALARY']].count()
```

## This is very important concept

```python
#df.loc[df['SALARY'].between(5000, 10000)]
df.loc[df['SALARY'].between(5000, 10000), ['EMAIL', 'SALARY']] # this is very very important concept
```

```python
# find out the employee first_name and last_name whose department_id in range 60 to  and also  sort the salary
df.loc[(df['DEPARTMENT_ID'].between(60,90)),['FIRST_NAME','LAST_NAME','SALARY']].sort_values('SALARY')
# here condition column is department_id
```

```python
df.loc[df['SALARY'].between(5000, 10000), ['FIRST_NAME', 'SALARY']] .sort_values('SALARY')
# here the condition column is salary
```

```python
#isin()--> Checks whether a value is one of several allowed values.
# alternative code without using isin ->
df[(df['JOB_ID']=='IT_PROG')|(df['JOB_ID'] == 'FI_MGR')]
```

```python
#show the details of employee whose job_id is 'IT_PROG'or 'FI_MGR'
df.loc[df['JOB_ID'].isin(['IT_PROG','FI_MGR'])]
```

```python
#show the details of employee whose job_id is 'IT_PROG'or 'FI_MGR' with only column of
#  'HIRE_DATE','JOB_ID','SALARY'
df.loc[df['JOB_ID'].isin(['IT_PROG','FI_MGR']),['HIRE_DATE','JOB_ID','SALARY']]
```

```python
#show the details of employee whose job_id is 'IT_PROG'or 'FI_MGR' with only column of
#  'HIRE_DATE','JOB_ID','SALARY' but salary should be sorted 
df.loc[(df['JOB_ID'].isin(['IT_PROG','FI_MGR']),['HIRE_DATE','JOB_ID','SALARY'])].sort_values('SALARY')
```

```python
# list the employee whose salary between 5000 to 10000 and who must be work in HR_REP job
df[(df['SALARY'].between(5000,10000)) & (df['JOB_ID']=='HR_REP')]
```

```python

# find employee who works under department 60 and 90 only
depart_under = (df['DEPARTMENT_ID']==60)|(df['DEPARTMENT_ID']==90)
df[depart_under]
```

# using both isin and between both function in same query

```python
# above same thingdone by using isin function
df[df['DEPARTMENT_ID'].isin([60,90])]
```

# this logic is very very important

```python
# find employee who works under department 60 and 90 and their salary must in range between(6000,10000)
emp_in = df['DEPARTMENT_ID'].isin([60,90]) & df['SALARY'].between(6000,10000)
df[emp_in]
```

```python
# find the employee detail who work in HR_REP job_id and salary between range between 5000,8000
new_emp = df['JOB_ID'].isin(['HR_REP']) & df['SALARY'].between(5000,8000)
df[new_emp]
```
